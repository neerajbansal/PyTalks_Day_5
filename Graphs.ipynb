```python
import pandas as pd
```

```python
graph_data = pd.read_excel('line_graph.xlsx')
print(graph_data)
```

```python
x_axis = list(graph_data['Year'])
y_axis = list(graph_data['Cars Sold'])

print("x axis data is %s and length is %s" %(x_axis, len(x_axis)))
print("y axis %s and length is %s" %(y_axis, len(y_axis)))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel('Year')
plt.ylabel('Number of Cars Sold.')
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis, autopct='%1.1f')
plt.show()
```

```python
plt.savefig('pie_chart.png')
```
